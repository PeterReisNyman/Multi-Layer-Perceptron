# Multi-Layer Perceptron

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score

import pandas as pd
import matplotlib as plt
from pandas.plotting import scatter_matrix

ds = sklearn.datasets.load_wine(return_X_y=False, as_frame=False)
ix = [i for i,d in enumerate(ds['feature_names']) if d =='od280/od315_of_diluted_wines']
ds['feature_names'][ix[0]] = 'ratio_of_diluted_wines'

X = ds['data']
Y = ds['target']
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.2)

plt.rcParams.update({
    "xtick.labelsize": 6,
    "ytick.labelsize": 6
})
```

```python
model_default = MLPClassifier(max_iter = 1000)
model_default.fit(X_train, Y_train)
model_default.n_layers_

predictions_default = model_default.predict(X_test)

print(f'{Y_test}\n\n{predictions_default}\n\n{accuracy_score(Y_test, predictions_default)}\n')

predictions_defaultTr = model_default.predict(X_train)

print(f'{Y_train}\n\n{predictions_defaultTr}\n\n{accuracy_score(Y_train, predictions_defaultTr)}')
```
